```python
#importing datetime module
import datetime 
#user input of card holder name
card_holder_name=input("Enter Card Holder Name")
#user input of card number
card_number=input("Enter your card number-: ")
flag=1
#user input of card expiry date in valid formate
while(1):
    try:   
        card_expiry_date=input("What is your card expiry date? (in DD/MM/YYYY)-: ")
        card_expiry_date=datetime.datetime.strptime(card_expiry_date,"%d/%m/%Y").date()
        flag=1
    except:
        print("Erro :Invalid Date Entered or Invalid Date Format")
        print("----Enter Again----")
        flag=0
    if flag==1:
        break
#user input of card number
cv=input("Enter CV number-: ")

#cheching the validation of card number by luhn algorithm.
def valid_sum(card_number):
    card_number_list=[]
    for i in card_number:
        card_number_list.append(int(i))
    sum=0
    for i in range(len(card_number_list)-2,-1,-2):
        card_number_list[i]=card_number_list[i]*2
        if card_number_list[i]>9:
            card_number_list[i]=card_number_list[i]%10+1
    sum=0
    for i in range(len(card_number_list)-1,-1,-1):
        sum=sum+card_number_list[i] 
    return sum%10==0

#checking the valid starting of card number
def valid_starting(card_number):
    return (card_number[0]=="4" or card_number[0]=="5" or card_number[0]=="6" or card_number[0]+card_number[1]=="37" )

#checking valid length of card
def valid_length(card_number):
    return len(card_number)>12 and len(card_number)<17

#checking valid expiry date
def valid_date(card_expiry_date): 
    today=datetime.date.today()
    return card_expiry_date>today

#checking valid cv
def valid_cv(cv):
    return (3 or 4)==len(cv)


#------------------------------Card Validation----------------------------
def is_valid_card():
    if valid_length(card_number) and valid_cv(cv) and valid_date(card_expiry_date) and valid_starting(card_number) and valid_sum(card_number):
        print("----Valid Card----")
    else:
        print("-----Invalid Card----")

#calling the is_valid_card function
is_valid_card()
```
